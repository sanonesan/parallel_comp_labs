```python
import numpy as np
import pandas as pd

out_dir = "./output/test0_for_approximation_order/"
file_name_2 = "test0_tau_1e-2"
file_name_3 = "test0_tau_1e-3"
file_name_4 = "test0_tau_1e-4"

def table(df: pd.DataFrame, err, p):

    pass


def u_func(t):
    return np.exp(np.sin(t)) - np.cos(t) + 0.5


def make_df(tabledata) -> pd.DataFrame:
        df = pd.DataFrame(tabledata, columns = [r'tau', 'AbsErr(tau)', 'delta', 'ln q delta']) 
        return df

def make_sheets_abs_err(u, y, time, tau, tabledata = []):

    er0 = abs(u(time) - y[0])
    er1 = abs(u(time) - y[1])
    er2 = abs(u(time) - y[2])
    er3 = 1

    er = [er0, er1, er2, er3]        
    for i in range(0,3):
        temp = [tau[i], er[i],  er[i]/ er[i+1], np.emath.logn(0.25, er[i]/ er[i+1])] 
        tabledata.append(temp)

    return tabledata

tau = [0.1, 0.05, 0.025]
```
